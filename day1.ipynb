```python
#Section 1 working with numpy
import numpy as np 
#create arrays
a=np.array([1,2,3])
b=np.arange(10)
c=np.random.rand(3,4)
print("Array a: ",a)
print("Array b (0 to 9): ",b)
print("Random 3x4 array c:\n",c)
```

```python
#array operation
print("a + 2:",a+2)
print("Mean of b:",np.mean(b))
print("Max value in c:",np.max(c))
```

```python
# Indexing and slicing
print("First row of c:", c[0])
print("Element at pos [1,2]:", c[1][2])
print("Subarray of b (2 to 5):", b[2:6])
```

```python
#Section 2 pandas
import pandas as pd
#create dataframe
data={
    'Name':['Alice','Bob','Charlie'],
    'Age':[25,30,35],
    'Score':[85.5,89.5,90.5]
}
df=pd.DataFrame(data)
print(df)
```

```python
#Inspecting data
print("DataFrames:\n", df)
print("Summary statistics:\n", df.describe())
print("Average Score:", df['Score'].mean())
```

```python
#Data Selection and filtering
print("People older than:28\n",df[df['Age']>28])
print("Names Column:\n",df['Name'])
```

```python
#Section 3 Matplotlib
import matplotlib.pyplot as plt
#line plot
x=np.linspace(0,10,100)
y=np.sin(x)

plt.figure(figsize=(8,4))
plt.plot(x,y)
plt.title("Sine wave")
plt.xlabel("x-axis")
plt.ylabel("sin(x)")
plt.grid(True)
plt.savefig("sine_wave_plot.png")
plt.show()
```

```python
#Histogram
data=np.random.randint(1000)
plt.figure(figsize=(6,4))
plt.hist(data,bins=30,color='green',edgecolor='black')
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.savefig("histogram.png")
plt.show()
```

```python
#Bonus Iris dataset analysis
from sklearn.datasets import load_iris
iris=load_iris()
df_iris=pd.DataFrame(data=iris.data,columns=iris.feature_names)
df_iris['target']=iris.target

print("Main feature values by target:\n",df_iris.groupby('target').mean)
pd.plotting.scatter_matrix(df_iris,figsize=(10,10),c=df_iris['target'],cmap='viridis')
plt.show()
```
